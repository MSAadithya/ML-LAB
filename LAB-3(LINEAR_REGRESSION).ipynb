```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/canada_per_capita_income.csv")

# Convert to numpy arrays
x = df['year'].values
y = df['per capita income (US$)'].values

n = len(x)

# Calculate slope (β1)
numerator = n * np.sum(x*y) - np.sum(x) * np.sum(y)
denominator = n * np.sum(x*x) - (np.sum(x))**2

b1 = numerator / denominator

# Calculate intercept (β0)
b0 = np.mean(y) - b1 * np.mean(x)

print("Intercept (β0):", b0)
print("Slope (β1):", b1)

# Predictions
y_pred = b0 + b1 * x

# Plot
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Linear Regression (Without Library)")
plt.show()
```

multi linear regression

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("hiring.csv")

# Convert words to numbers
df['experience'] = df['experience'].replace({
    'two': 2,
    'three': 3,
    'five': 5,
    'seven': 7,
    'ten': 10,
    'eleven': 11
})

df['experience'] = df['experience'].fillna(0)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())

# Feature matrix
X = df[['experience','test_score(out of 10)','interview_score(out of 10)']].values
Y = df['salary($)'].values

# Add column of 1s for intercept
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))

# Matrix formula
Xt = X.T
beta = np.linalg.inv(Xt.dot(X)).dot(Xt).dot(Y)

print("Beta values (Intercept, b1, b2, b3):")
print(beta)

# Prediction example (2 yrs exp, 9 test, 6 interview)
new_data = np.array([1, 2, 9, 6])
prediction = new_data.dot(beta)

print("Predicted salary:", prediction)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("/content/salary.csv")

df.fillna(df.mean(), inplace=True)

# Convert to numpy arrays
x = df['YearsExperience'].values
y = df['Salary'].values

n = len(x)

# Calculate slope (β1)
numerator = n * np.sum(x*y) - np.sum(x) * np.sum(y)
denominator = n * np.sum(x*x) - (np.sum(x))**2

b1 = numerator / denominator

# Calculate intercept (β0)
b0 = np.mean(y) - b1 * np.mean(x)

print("Intercept (β0):", b0)
print("Slope (β1):", b1)

# Predictions
y_pred = b0 + b1 * x

# Plot
plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, color='red')
plt.xlabel("Year")
plt.ylabel("Per Capita Income")
plt.title("Linear Regression (Without Library)")
plt.show()
```

Multiple LR with library ( 1000_company dataset)

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/1000_Companies.csv")

# One Hot Encoding for State
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Features and Target
X = df.drop("Profit", axis=1)
y = df["Profit"]

# Model
model = LinearRegression()
model.fit(X, y)

# Results
print("Intercept:", model.intercept_)
print("Coefficients:")
for feature, coef in zip(X.columns, model.coef_):
    print(feature, ":", coef)

# Example prediction
sample = [[160000,130000,300000,1,0]]  # Adjust based on encoded columns
print("Predicted Profit:", model.predict(sample))
```

## Multiple LR ( without library) ( 1000 companies)

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/1000_Companies.csv")

# One Hot Encoding
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Separate features and target
X = df.drop("Profit", axis=1).values
Y = df["Profit"].values

# Add column of 1s (Intercept)
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X)).astype(float) # Explicitly cast X to float

# Normal Equation
Xt = X.T
beta = np.linalg.inv(Xt.dot(X)).dot(Xt).dot(Y)

print("Beta values (Intercept first):")
print(beta)

# Example prediction
sample = np.array([1,160000,130000,300000,1,0])
prediction = sample.dot(beta)

print("Predicted Profit:", prediction)
```
